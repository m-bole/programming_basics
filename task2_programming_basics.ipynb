```python
#Using input functions and conditions, create a simple “calculator”.
#The program should allow the user to enter two numbers and information
#about an operation of their choice to perform on them (3 pts).


def calculator():
    """A calculator that allows user to enter two numbers and an operation of their choice"""

    print("Hi! I am a calculator. I will calculate something for you.")

    while True:
        try:
            number1 = float(input("Enter first number: "))
            number2 = float(input("Enter second number: "))
            operation = input("Enter operation (+, -, *, /): ")

            if operation == "+":
                result = number1 + number2
            elif operation == "-":
                result = number1 - number2
            elif operation == "*":
                result = number1 * number2
            elif operation == "/":
                if number2 == 0:
                    print("Error: Division by zero!")
                    continue
                result = number1 / number2
            else:
                print("Invalid operation. Please try again.")
                continue

            print(f"Result: {result}")

            another = input("Do you want to perform another calculation? (yes/no): ").lower()
            if another != 'yes':
                print("Thank you for using me to calculate something. :) Have a nice day.")
                break

        except ValueError:
            print("Invalid input. Please enter valid numbers.")

calculator()
```

```python
def registration_form():
    """An enhanced prototype of a user registration form on the website"""

    print("Welcome to our registration form! Please follow the instructions!")

    def get_alpha_input(input):
        while True:
            value = input(input)
            if value.isalpha():
                return value
            else:
                print("Invalid input. Please use only letters.")

    def is_valid_password(password):
        return (len(password) >= 5 and
                any(char.isdigit() for char in password) and
                any(char.isupper() for char in password))

    def get_user_data():
        name = get_alpha_input("Enter your name: ")
        last_name = get_alpha_input("Enter your last name: ")

        while True:
            try:
                age = int(input("Enter your age: "))
                if age < 18:
                    print("Too young. User needs to be over 18 years old to register.")
                elif age > 120:
                    print("Invalid age. Please enter a valid age.")
                else:
                    break
            except ValueError:
                print("Invalid input. Please enter a number for age.")

        while True:
            password = input("Enter your password (at least 5 characters, 1 number, and 1 capital letter): ")
            if not is_valid_password(password):
                print("Invalid password. Please ensure it's at least 5 characters long, contains a number, and a capital letter.")
                continue

            repeat_password = input("Repeat your password: ")
            if password == repeat_password:
                break
            else:
                print("Passwords do not match. Please try again.")

        while True:
            phone_number = input("Enter your phone number (9 digits): ")
            if phone_number.isdigit() and len(phone_number) == 9:
                break
            else:
                print("Invalid phone number. Please enter a 9-digit number.")

        return name, last_name, age, password, phone_number

    while True:
        name, last_name, age, password, phone_number = get_user_data()

        print("\nRegistration Information:")
        print("Name:", name)
        print("Last Name:", last_name)
        print("Age:", age)
        print("Phone Number:", phone_number)

        # Password display options
        while True:
            pwd_option = input("How would you like to display the password? (full/hidden): ").lower()
            if pwd_option == "full":
                print("Password:", password)
            elif pwd_option == "hidden":
                print("Password:", "*" * len(password))
            else:
                print("Invalid option. Please choose 'full' or 'hidden'.")
                continue
            break

        confirmation = input("\nIs the form data correct? (yes/no): ").lower()
        if confirmation == "yes":
            print("Thank you for completing the form. :) Have a nice day.")
            break
        else:
            print("Let's fill the form again.")

registration_form()
```
